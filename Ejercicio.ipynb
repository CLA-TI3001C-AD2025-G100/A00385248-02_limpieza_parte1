## Ejercicio
Considerando el archivo 'data/Alarm Survey Data.xlsx', limpie los datos para que sean adecuados para análisis

Pasos:
1. Leer los datos
2. Meterlos en dataframe
3. Revisar los tipos de datos
4. Convertir las columnas que sean necesarias

```python
import pandas as pd
```

```python
# leemos el archivo 'data/Alarm Survey Data.xlsx' y lo guardamos en un DataFrame
df = pd.read_excel('data/Alarm Survey Data.xlsx')   
df
```

```python
df.dtypes
```

```python
# quitar de la columna alarm_rating de df la cadena ' stars'
df['alarm_rating'] = df['alarm_rating'].str.replace(' stars', '', regex=False)
df
```

```python
df.dtypes
```

```python
# convertir la columna alarm_rating a tipo numerico
df['alarm_rating'] = pd.to_numeric(df['alarm_rating'], errors='coerce')
df
```
